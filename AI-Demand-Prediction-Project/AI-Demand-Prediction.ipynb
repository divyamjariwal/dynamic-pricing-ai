```python
import pandas as pd
import numpy as np

np.random.seed(42)

# CONFIG
num_products = 20
rows_per_product = 1000

# REAL PRODUCT NAMES
product_names = [
    "Wireless Earbuds", "Gaming Laptop", "Smartphone", "Bluetooth Speaker",
    "Smartwatch", "Tablet", "Noise Cancelling Headphones", "Power Bank",
    "DSLR Camera", "Fitness Tracker", "LED Monitor", "Mechanical Keyboard",
    "Wireless Mouse", "Inkjet Printer", "External Hard Drive",
    "Gaming Console", "Portable SSD", "Router", "USB Hub", "Smart Home Speaker"
]

product_ids = [f"P{i+1}" for i in range(num_products)]

data = []

# DATE RANGE
dates = pd.date_range(start="2023-01-01", end="2024-12-31")

# GENERATION LOOP
for i in range(num_products):

    # product-specific behavior
    base_price = np.random.uniform(50, 800)
    demand_sensitivity = np.random.uniform(0.5, 1.5)

    for _ in range(rows_per_product):

        date = np.random.choice(dates)
        day = pd.to_datetime(date).day_name()

        # discount
        discount = np.random.choice([0, 5, 10, 15, 20])

        # price variation
        price = base_price * (1 - discount/100) + np.random.uniform(-15, 15)

        # competitor pricing
        competitor_price = price + np.random.uniform(-30, 30)

        # ratings & sentiment
        rating = np.random.uniform(3.5, 5.0)
        sentiment = (rating - 3.5)/1.5 + np.random.uniform(-0.1, 0.1)

        # time effect
        weekend_boost = 20 if day in ["Saturday", "Sunday"] else 0

        # seasonal effect
        month = pd.to_datetime(date).month
        seasonal_boost = 25 if month in [10, 11, 12] else 0

        # demand formula (balanced)
        demand = (
            350
            - demand_sensitivity * price
            + 4 * discount
            + 50 * sentiment
            + weekend_boost
            + seasonal_boost
            + np.random.normal(0, 20)
        )

        demand = max(5, int(demand))

        data.append([
            date,
            product_ids[i],
            product_names[i],
            round(price, 2),
            demand,
            day,
            discount,
            round(rating, 2),
            round(sentiment, 2),
            round(competitor_price, 2)
        ])

# CREATE DATAFRAME
df = pd.DataFrame(data, columns=[
    "date", "product_id", "product_name", "price",
    "demand", "day_of_week", "discount",
    "rating", "sentiment_score", "competitor_price"
])

# SAVE DATASET
df.to_csv("data/raw/dynamic_pricing_dataset.csv", index=False)

print("Dataset created successfully!")
print("Shape:", df.shape)
```
